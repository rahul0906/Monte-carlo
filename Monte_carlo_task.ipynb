## **Monte Carlo Simulator**
You have to sample different paths using simulations and store them in a 2D list of size (max_simulations*(timesteps + 1))

timesteps + 1 is due to the initial element being the starting price of day 0

```python
timesteps = 30
starting_price = 600
volatility = 0.02
max_simulations = 5000

final_list = []
# modeling a random price walk over 30 days
def randomWalk(vol, ini_price):
    prices = [ini_price]
    price = ini_price
    for i in range(1,timesteps+1):
        price = price + price*random.normal(0,vol)
        prices.append(price)
    return(prices)

# Conduct calculation and building 2D resultant matrix
for i in range(0,max_simulations+1):
    prices = []
    prices = randomWalk(volatility,starting_price)
    final_list.append(prices)

# Creading final_values list
final_values = []
for i in final_list:
    final_values.append(i[-1])
```
